```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default="plotly_white"
```

```python
df=pd.read_csv("customer_acquisition_data.csv",encoding='latin-1')
```

```python
df
```

```python
fig=px.histogram(df,x='cost',nbins=20,
                title='Distribution of Acquistion Cost')
fig.show()
```

```python
cost_by_channel=df.groupby('channel')['cost'].mean().reset_index()
cost_by_channel
```

```python
fig=px.bar(cost_by_channel,x="channel",y='cost',
          title='Cost of channel')
fig.show()
```

```python
df
```

```python
conversion_by_channel=df.groupby('channel')['conversion_rate'].mean().reset_index()
fig=px.bar(conversion_by_channel,x='channel',y='conversion_rate',title='conversion rate by channel')
fig.show()
```

```python
df
```

```python
total_revenue_channel=df.groupby('channel')['revenue'].sum().reset_index()
total_revenue_channel
```

```python
fig=px.pie(total_revenue_channel,values='revenue',names='channel',title='Total Revenue of Most profitable Channel',hole=0.5)
fig.show()
```

```python
df['rio']=df['revenue']/df['cost']
```

```python
df
```

```python
total=df.groupby('channel')['rio'].mean().reset_index()
total
```

```python
fig=px.bar(total,x='channel',y='rio',title='Return on Inverstment (RIO) by channel')
fig.show()
```
